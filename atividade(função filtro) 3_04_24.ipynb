```python
import pandas as pd
```

```python
data = {
    'Nome': ['Amaciante', 'Alvejante', 'Sabão em pó', 'Sabão liquído'],
    'Quantidade': [10, 20, 100, 200]
}

df = pd.DataFrame(data)
print(df)
```

```python
filtro = df[df['Quantidade'] > 50]
print(filtro)
```
