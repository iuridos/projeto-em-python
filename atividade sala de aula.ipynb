```python
data = {
     "Nome": ["Controle", "Sofá", "Mesa", "PS5", "Guarda-Roupa"],
     "Categoria": ["Eletrônicos", "Movél", "Movél", "Eletrônicos", "Movél"],
     "Preços": [20, 1000, 800, 3000, 80],
     "Quantidade": [10, 20, 5, 7, 8]
}

df = pd.DataFrame(data)
print(df)
```

```python
filtro = (df['Categoria']=="Eletrônicos")&(df['Preços'] <100)
resultado = df[filtro]
print(resultado)
```

```python
df.loc[df['Preços'] < 100, 'Barato ou Caro'] = 'Barato'
df.loc[df['Preços'] > 100, 'Barato ou Caro'] = 'Caro'
df
```
