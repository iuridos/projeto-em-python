```python
import pandas as pd
```

```python
data = {
    'Salário':[3000, 5000, 3000, 4000, 20000, 15000, 3000]
}

df = pd.DataFrame(data)
```

```python
media = df['Salário'].mean()
print(media)
```

```python
mediana = df['Salário'].median()
print(mediana)
```

```python
moda = df['Salário'].mode().iloc[0]
print(moda)
```

```python
data = {
    'Vendas':[10, 20, 30, 40, 50],
    'Preços':[100, 120.25, 250.75, 150.68, 980,74]
}

df = pd.DataFrame(data)
```

```python
media = df['Vendas'].mean()
print(media)
```

```python
mediana = df['Vendas'].median()
print(mediana)
```

```python
moda = df['Vendas'].mode()[0]
print(moda)
```
