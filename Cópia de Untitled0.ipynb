```python
#receba 4 notas de um aluno e exibe se ele foi aprovado(nota maior que 6)
#se ele ficou de recuperação(nota maior ou igual a 4) ou se ele foi
#reprovado (nota menor que 4)
```

```python
nome= input("digite seu nome: ")

nota1=float(input('digite a nota 1: '))
nota2=float(input('digite a nota 2: '))
nota3=float(input('digite a nota 3: '))
nota4=float(input('digite a nota 4: '))


```

```python
#Dividindo as notas pela quantidade de notas
media = (nota1 + nota2 + nota3 + nota4) /4

if media >=6:
   print(f'Parabéns {nome}, voce foi APROVADO com a nota{media}!')
elif media >=4:
   print(f'Você tirou a nota {media}, Nos vemos na RECUPERAÇÃO, {nome}!')
else:
   print(f'Infelizmente você foi Reprovado com a nota {media},{nome}, Nos vemos no próximo ano!')
```
