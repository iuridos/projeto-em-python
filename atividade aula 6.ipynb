```python
import pandas as pd
```

```python
Remédios = {
    'Remédios':['Dipirona', 'amoxicilina', 'Doril', 'Diclofenaco', 'Piroxicam'],
    'Quantidade': [20, 33, 66, 9, 10],
    'Valor': [3, 6, 20, 30, 45]
}

df = pd.DataFrame(Remédios)
print(df)
```

```python
filtro = df[(df['Quantidade'] < 10) & (df['Valor'] > 20)]
print('\n PRODUTOS ABAIXO DA QUANTIDADE E PREÇO ACIMA DE 20')
print(filtro)
```

```python
História = {'Alunos':['Iuri', 'Amanda', 'Alice', 'Raimunda', 'Letícia']}

df = pd.DataFrame(História)
print(df)
```

```python
filtro = df[df["Alunos"].str.contains("^A")]
print('\nNOME DOS ALUNOS QUE COMEÇAM COM A LETRA (A)')
print(filtro)
```
